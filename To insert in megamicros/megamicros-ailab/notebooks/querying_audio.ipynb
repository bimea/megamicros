## Extraire des signaux annotés
Il existe de nombreuses façons d'accéder aux signaux enregistrés dans la base. Si la base est annoté, alors il est possible d'accéder aux données audio annotées par les labels qui leur ont été affectés.
Il faut pour cela connaître les labels.

Pour obtenir le domaine d'application et l'ensemble des labels, reportez vous au Notebook *querying_metadata.ipynb*.

Nous choisissons le domaine applicatif numéro 1 (bien-être animal) et le label d'identifiant 15 (code: *piglet-squeals*).
Il s'agit de hurlements de porcelets susceptibles de subir un écrasement. Pour récupérer des signaux de la base étiquetés avec ce label nous utilisons la méthode *load_labeling* en limitant à 10 le nombre maximum d'entités (*limit=10*). 
Dans l'exemple ci-dessous nous récupérons les voies des microphones 1, 2, 3, 5, 6, 7:

````python
label_id = 15
limit=10
channels=[1,2,3,5,6,7]

# chargement des signaux
with AidbSession( dbhost='http://dbwelfare.biimea.io/', login='ailab', email='bruno.gas@biimea.com', password='#T;uZnQ5UJ_JC~&' ) as session:
    audio = session.load_labelized( label_id=label_id, limit=limit, channels=channels )

# Affichage
print( f"{len(audio)} section audio récupérées:" )
for idx, aud in enumerate( audio ):
    print( f"Audio[{idx}]: {aud}")
````

## Différences inter-voies

````python
s1 = audio[0].channel(0)
s2 = audio[0].channel(1)
N = audio[0].samples_number
sf = audio[0].sampling_frequency
t = np.arange( N ) / sf

figure, axes = plt.subplots(2, 1)
offset = 10000
size = 256
axes[0].plot(t, s1, t, s2)
axes[1].plot(t[offset:offset+size], s1[offset:offset+size], t[offset:offset+size], s2[offset:offset+size])
````
