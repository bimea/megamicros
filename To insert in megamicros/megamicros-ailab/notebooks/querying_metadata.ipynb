## Recherche des domaines
Une base peut comporter des données se référent à plusieurs domaines d'application différents. 
Vous pouvez connaître la liste des domaines et récupérer lm'identifiant du domaine qui vous interesse:

````python
domains = session.load_domains()
print( "domains=", domains )
````

## Obtention des labels
Pour connaître l'ensemble des labels définis pour sur un champ applicatif donné, utilisez la méthode *load_labels()*.
Cette méthode retourne le nom, le code, l'identifiant et la date de création pour tous les labels définis:

````python
labels = session.load_labels( domain_id=1 )
print( "labels=", labels )
````
