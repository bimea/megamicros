# Getting started with MemsArrayNP objects

The `MemsArrayNP` derived from the `MemsArray` class while being very similar.
While `MemsArray` objects cannot generate any signals except random ones,`MemsArrayNP` can get their signals from numpy array data.

```python
import numpy as np
import matplotlib.pyplot as plt
from megamicros_tools.log import log
from megamicros.core.np import MemsArrayNP

log.setLevel( "INFO" )
```

Let's create an array of random data aiming at simulate signals from a 32 MEMs antenna.

```python
data = np.random.rand( 10000, 32 )
data.shape
```

Create the antenna

```python
antenna = MemsArrayNP( data )
```

Specify avalable the 32 MEMs channels and activate them:

```python
antenna.setAvailableMems( 32 )
antenna.setActiveMems( [i for i in range(32)] )
```

Run

```python
antenna.run( duration=2 )

# get data, loop is blocking
for data in antenna:
    print( f"data={data}" )

# Wait until the thread terminates
antenna.wait()
```
