# Getting started with MemsArrayCB objects

The `MemsArrayCB` derived from the `MemsArray` class while being very similar.
While `MemsArray` objects cannot generate any signals except random ones,`MemsArrayCB` objects can get their signals from result of a user defined callback function.

```python
import numpy as np
import matplotlib.pyplot as plt
from megamicros_tools.log import log
from megamicros.core.cb import MemsArrayCB

log.setLevel( "INFO" )
```

Let's create a callback function aiming at simulate random signals from a 32 MEMs antenna.

```python
def my_callback_generator():
    data = np.random.rand( 256, 32 )
    return data
```

Create the antenna

```python
antenna = MemsArrayCB( user_in_callback=my_callback_generator, user_in_callback_args=[] )
```

Run on two seconds iterations

```python
antenna.setAvailableMems( 32 )
antenna.run( duration=2, datatype='float32', frame_length=256, mems=[i for i in range(32)] )

# get data, loop is blocking
for data in antenna:
    print( f"data={data}" )

# Wait until the thread terminates
antenna.wait()
```
