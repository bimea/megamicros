### Informations

The ``AidbSession.getSourcefileInfo()`` method allows you to get detailed informations about an audio file. 
The file reference can be its DB identifier, or its url address, or its file name. 

```python
with AidbSession(
    dbhost='http://dbwelfare.bimea.tech/',
    login='ailab',
    email='bruno.gas@biimea.com',
    password='#T;uZnQ5UJ_JC~&' ) as session:

        fileinfo = session.getSourcefileInfo( id=2 )
        print( "file info:" )
        pprint( fileinfo )     

        fileinfo = session.getSourcefileInfo( url='http://dbwelfare.biimea.io/sourcefile/10/' )
        print( "file info:" )
        pprint( fileinfo )

        fileinfo = session.getSourcefileInfo( filename='mu5h-20220710-225910.h5' )
        print( "file info:" )
        pprint( fileinfo )
```

### Content

The ``downloadSourcefile`` allow to download file content from the database

```python
with AidbSession(
    dbhost='http://dbwelfare.bimea.tech/',
    login='ailab',
    email='bruno.gas@biimea.com',
    password='#T;uZnQ5UJ_JC~&' ) as session:
        file_info = session.getSourcefileInfo( id=2 )
        file_content = session.downloadSourcefile( id=2 )

        h5filename = file_info['H5 filename']
        print( f"save {h5filename} file" )
        with open( h5filename, 'bw' ) as file:
            file.write( file_content )

        file_content = session.downloadSourcefile( id=2, as_wav=True )
        wavfilename = file_info['WAV filename'].replace(".h5", ".wav")
        print( f"save {wavfilename} file" )
        with open( wavfilename, 'bw' ) as file:
            file.write( file_content )
```
